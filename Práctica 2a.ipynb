# Histograma con OpenCV

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('ImagenRGB.jpg', cv2.IMREAD_GRAYSCALE)
cv2.imshow('ImagenRGB.jpg', img)

hist = cv2.calcHist([img], [0], None, [256], [0, 256])
plt.plot(hist, color='gray' )

plt.xlabel('Valores de intensidad')
plt.ylabel('NÃºmero de pixeles')
plt.show()

cv2.destroyAllWindows()
```

# Histograma con Pillow

```python
from PIL import Image

#modulo para generar la grafica
import matplotlib.pyplot as plt

def graficar(datos, nombre_del_archivo):
   
    plt.figure(1)
    x=range(len(datos))
    plt.xticks([0, 50, 100, 150, 200, 255],[0, 50, 100, 150, 200, 255])
    plt.bar(x, datos, align='center')
    plt.title('Histograma')
    plt.xlabel('Valores de intensidad')
    plt.ylabel('Numero de pixeles')
    
    plt.savefig(nombre_del_archivo, bbox_inches='tight')

    return None

foto=Image.open('ImagenRGB.jpg')

#si la imagen es a color la convertimos a escala de grises
if foto.mode != 'L':
    foto=foto.convert('L')

histograma=foto.histogram()

graficar(histograma, 'HistogramaPIL.jpg')
```
